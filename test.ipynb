```python
from utils import *
from matplotlib import rc_file_defaults
import matplotlib.pyplot as plt
```

```python
def plot_mean_param_evolution(thetas,num_groups, title_ = 'Group-wise Mean Parameter Evolution'):
    thetas1 = np.mean(thetas, axis=0)
    plt.rc('text', usetex = True)
    plt.rc('font', family='serif')
    width = 6
    height = 4
    fig, ax = plt.subplots(nrows=1, ncols=1) 
    fig.set_size_inches(width, height) #exact size of the figure
    fig.set_dpi(250) 
    lw = 1
    colors = ['#DBB40C','red','blue','tab:green']

    for i in range(num_groups):
        ax.plot(thetas1[:,i], '--', label='Group '+str(i+1), linewidth=lw, c=colors[i])

    ax.set_title(title_)
    ax.grid()
    ax.set_xlabel('Rounds')
    ax.set_ylabel('Proportion')
    ax.legend()
    plt.show()
```

```python
num_rounds = 500
lambda_ = 3
eta = 0.025
N = 50
num_instances = 50
fairness_target = np.array([0.4, 0.3, 0.2, 0.1])
initial_theta = np.array([0.1, 0.2, 0.3, 0.4])
num_groups = len(fairness_target)
mean = np.ones(num_groups)*5
std = np.ones(num_groups)
acceptance_ratio = 0.6
decay = 0

applicants, admitteds, thetas = process_multigroup(N, mean, std, acceptance_ratio, fairness_target, initial_theta, lambda_, eta, num_instances, num_rounds, decay)
plot_mean_param_evolution(thetas,num_groups)
```
